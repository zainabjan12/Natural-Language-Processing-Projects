# Analyzing British Airline Reviews for Service Improvement

In this project, I conducted an extensive analysis of British airline reviews using Python and Natural Language Processing (NLP) techniques. The primary goal was to extract actionable insights from customer feedback to drive improvements across various service aspects within the airline industry. Here's an overview of the project's key components:

#### 1.  Data Acquisition and Preprocessing:
Used British airline reviews dataset from Kaggle which was extracted from https://www.airlinequality.com/airline-reviews/british-airways.
Implemented robust data preprocessing procedures, including data cleaning, duplicate removal, handling missing values, and text normalization. These steps were crucial to ensure data readiness for analysis.

#### 2. Feature Engineering:
I improved the dataset by adding features for better customer sentiment analysis. This included extracting year and month from dates and creating an 'is_verified' column based on the reviews.

#### 3. Exploratory Data Analysis (EDA):
Conducted Exploratory Data Analysis (EDA) to gain a holistic understanding of data distribution.
Utilized various visualization techniques, such as word clouds, bar charts, and line plots, to identify trends and visualize essential patterns within the dataset.

#### 4. Topic Modeling:
I conducted manual topic identification and categorization of prevalent themes in the reviews."

#### 5. Categorization based on Keywords:
I categorized reviews using predefined keywords and explored various NLP techniques like NMF and LDA. Ultimately, I opted for the keyword-based approach due to its effectiveness

#### 6. Category Analysis:
Delved into a comprehensive analysis of each review category, enabling a profound understanding of the sentiment, frequency, and severity of issues mentioned within each group.

#### 7. Text-based Summarization:
Applied text summarization techniques to generate concise and informative summaries for each review category.
These summaries served as quick reference points, highlighting the central themes and insights within each set of reviews.

###### This project showcases the application of NLP and Python in extracting actionable insights from customer feedback to enhance various aspects of service delivery within the airline industry.

### Supported Functions

```python
#This function will create bargraphs for specific column for top 10 records
def visualize_top_categorical(data, column_name, top_n=10):
    # Count the frequency of each category
    category_counts = data[column_name].value_counts()

    # Select the top N categories
    top_categories = category_counts.head(top_n)

    # Create a color map for the bars
    colors = plt.cm.viridis(np.linspace(0, 1, len(top_categories)))  # Change colormap as needed

    # Calculate percentages
    total_count = len(data)
    percentages = (top_categories / total_count) * 100

    # Create a bar plot with colorful bars
    plt.figure(figsize=(10, 6))
    bars = plt.bar(range(len(top_categories)), top_categories, color=colors)

    # Add data labels as percentages above the bars
    for i, (bar, percent) in enumerate(zip(bars, percentages)):
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 10, f'{percent:.2f}%', ha='center', va='bottom', fontsize=12)

    plt.title(f'Top {top_n} {column_name} Distribution')
    plt.xlabel(column_name)
    plt.ylabel('Frequency')
    plt.xticks(range(len(top_categories)), top_categories.index, rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
# this function will plot grouped bargraph for each column based on recommended column
def visualize_grouped_categorical_percentage(data, group_column, category_column, top_n=10):
    # Filter the top N categories in the grouping column
    top_categories = data[group_column].value_counts().head(top_n).index

    # Initialize lists to store data
    yes_percentages = []
    no_percentages = []

    # Calculate percentages for 'Yes' and 'No' for each category
    for category in top_categories:
        subset = data[data[group_column] == category]
        total_count = len(subset)
        yes_count = (subset[category_column] == 'yes').sum()
        no_count = total_count - yes_count

        yes_percentage = (yes_count / total_count) * 100
        no_percentage = (no_count / total_count) * 100

        yes_percentages.append(yes_percentage)
        no_percentages.append(no_percentage)

    # Create the grouped bar chart
    plt.figure(figsize=(12, 6))
    x = np.arange(len(top_categories))
    width = 0.35

    plt.bar(x - width/2, yes_percentages, width, label='Yes', color='skyblue', alpha=0.7)
    plt.bar(x + width/2, no_percentages, width, label='No', color='salmon', alpha=0.7)

    # Set x-axis labels as the categories
    plt.xticks(x, top_categories, rotation=45, ha='right')

    # Add percentage labels above the bars
    for i, (yes_percentage, no_percentage) in enumerate(zip(yes_percentages, no_percentages)):
        plt.text(x[i] - width/2, yes_percentage + 1, f'{yes_percentage:.1f}%', ha='center', va='bottom', fontsize=10)
        plt.text(x[i] + width/2, no_percentage + 1, f'{no_percentage:.1f}%', ha='center', va='bottom', fontsize=10)

    plt.title(f'Percentage Distribution of {category_column} by {group_column}')
    plt.xlabel(group_column)
    plt.ylabel('Percentage (%)')
    plt.legend(title=category_column)
    plt.tight_layout()
    plt.show()
```

##  Feedback Analysis

```python
#filtering negative feedbacks only
neg_df=df[df['recommended']=='no']
#comment out the below line of code if you want to analyze only verified reviews
#neg_df=neg_df[neg_df['is_verified']=='Yes'] 
neg_df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import nltk
import re
from nltk.util import ngrams
from collections import Counter
from nltk.corpus import stopwords
from wordcloud import WordCloud
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from sklearn.feature_extraction.text import TfidfVectorizer

# Function to analyze negative feedbacks
def analyze_negative_feedbacks(negative_df, text_column, max_features=100):
    # TF-IDF vectorization with stop words removal
    tfidf_vectorizer = TfidfVectorizer(max_features=max_features, stop_words='english')
    tfidf_matrix = tfidf_vectorizer.fit_transform(negative_df[text_column])
    tfidf_features = tfidf_vectorizer.get_feature_names_out()

    # Word Cloud generation
    text = negative_df[text_column].str.cat(sep=' ')
    for word in ['British', 'Airways', 'BA', 'ba', 'airline', 'said', 'Verified', 'Review','flight']:
        text = text.replace(word, '')  # Remove excluded words
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('Word Cloud of Common Issues (All Negative Feedback)')
    plt.show()

    # Sentiment Analysis
    analyzer = SentimentIntensityAnalyzer()
    sentiment_scores = []

    sentiment = analyzer.polarity_scores(negative_df[text_column].str.cat(sep=' '))
    sentiment_scores.append({
        'Sentiment_Scores': sentiment
    })

    # Print sentiment scores
    print(f"Sentiment Scores for All Negative Feedback: {sentiment_scores[0]['Sentiment_Scores']}")

# Function to extract and plot n-grams
def documentNgrams(documents, size):
    ngrams_all = []
    stop_words = set(stopwords.words('english'))  # Define stopwords
    excluded_words = ['British', 'Airways', 'BA', 'ba', 'airline', 'said', 'Verified', 'Review', 'flight',
                      '', 'business', 'class', 'ticket', 'one', 'get', 'would', 'even', 'though','London']
    
    for document in documents:
        tokens = document.split()
        tokens = [token for token in tokens if token.lower() not in stop_words and token not in excluded_words]  # Remove stopwords and excluded words
        if len(tokens) <= size:
            continue
        else:
            output = list(ngrams(tokens, size))
        for ngram in output:
            ngram_text = " ".join(ngram)
            ngrams_all.append(ngram_text)
    
    cnt_ngram = Counter()
    for word in ngrams_all:
        cnt_ngram[word] += 1
    
    df = pd.DataFrame.from_dict(cnt_ngram, orient='index').reset_index()
    df = df.rename(columns={'index':'words', 0:'count'})
    df = df.sort_values(by='count', ascending=False)
    df = df.head(15)
    df = df.sort_values(by='count')
    
    return df

# Function to plot n-grams
def plotNgrams(documents):
    unigrams = documentNgrams(documents, 1)
    bigrams = documentNgrams(documents, 2)
    trigrams = documentNgrams(documents, 3)
    
    # Set plot figure size
    fig = plt.figure(figsize=(20, 7))
    plt.subplots_adjust(wspace=.5)

    ax = fig.add_subplot(131)
    ax.barh(np.arange(len(unigrams['words'])), unigrams['count'], align='center', alpha=.5)
    ax.set_title('Unigrams')
    plt.yticks(np.arange(len(unigrams['words'])), unigrams['words'])
    plt.xlabel('Count')

    ax2 = fig.add_subplot(132)
    ax2.barh(np.arange(len(bigrams['words'])), bigrams['count'], align='center', alpha=.5)
    ax2.set_title('Bigrams')
    plt.yticks(np.arange(len(bigrams['words'])), bigrams['words'])
    plt.xlabel('Count')

    ax3 = fig.add_subplot(133)
    ax3.barh(np.arange(len(trigrams['words'])), trigrams['count'], align='center', alpha=.5)
    ax3.set_title('Trigrams')
    plt.yticks(np.arange(len(trigrams['words'])), trigrams['words'])
    plt.xlabel('Count')

    plt.show()

analyze_negative_feedbacks(neg_df, 'reviews', max_features=100)
plotNgrams(neg_df['reviews'])
```

## Categorization of Issues based on Keywords.

I conducted an extensive review categorization process using a keyword-based approach. This involved the utilization of predefined keywords to categorize reviews into distinct issue categories. The chosen method, which relied on keyword matching, was selected for its effectiveness in assigning reviews to specific issue categories.

```python
issues_keywords = {
    'Check-in & Boarding': ['check-in', 'boarding', 'ticketing', 'counter', 'boarding pass'],
    'Flight Delays & Cancellations': ['cancel','delay', 'cancellation', 'late', 'reschedule','missed','flight cancelled'],
    'Customer Service & Support': ['lack of compassion','lack of interest','bad','nightmare','terrible','shocking attitude','bad experience','service', 'staff', 'crew', 'hospitality', 'customer support', 'complaints','complained','complaint','complain'],
    'Baggage Issues': ['baggage', 'luggage', 'lost luggage', 'claim', 'delayed baggage','lost'],
    'Food & Catering': ['food', 'meal', 'catering', 'menu', 'meal quality', 'inadequate snacks','unappetizing','tasteless'],
    'Seating & Comfort': ['seating', 'seat', 'legroom', 'cabin', 'uncomfortable seating', 'limited legroom'],
    'Entertainment & Amenities': ['entertainment', 'in-flight movies', 'TV', 'music', 'amenities'],
    'Cleanliness & Hygiene': ['cleanliness', 'hygiene', 'sanitation', 'dirty'],
    'Security & Safety': ['security', 'safety', 'screening', 'checks'],
    'Booking & Refund Issues': ['booking', 'reservation', 'website', 'confirmation', 'refund', 'voucher', 'booking difficulties', 'refund problems'],
}

# Create a new column to store the assigned issue, if text doesnt contain keywords then assign the issue Other
neg_df['Assigned_Issue'] = 'Other'

# Iterate through reviews and assign issues based on keyword matching
for index, row in neg_df.iterrows():
    review = row['reviews']
    assigned_issue = None
    for issue, keywords in issues_keywords.items():
        if any(keyword in review.lower() for keyword in keywords):
            assigned_issue = issue
            break  # Stop checking once a match is found
    if assigned_issue is not None:
        neg_df.at[index, 'Assigned_Issue'] = assigned_issue
```

```python
def generate_pie_chart(data, column_name, title):
    # Count the occurrences of each unique value in the specified column
    value_counts = data[column_name].value_counts()

    # Get the labels for the pie chart
    labels = value_counts.index

    # Create a pie chart without data labels and percentages
    plt.figure(figsize=(8, 8))
    plt.pie(value_counts, labels=None, startangle=140)

    # Create legends with labels
    plt.legend(labels, loc='upper right', bbox_to_anchor=(1, 0.9))

    plt.title(title)

    # Display the pie chart
    plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular
    plt.show()

generate_pie_chart(neg_df, 'Assigned_Issue', 'Assigned Issues')
```

## Trends Analysis for each category

```python
def generate_star_rating_line_plots(data, issue_column, year_column, rating_column):
    # Assuming 'date' is your date column, convert it to a datetime object
    data['date'] = pd.to_datetime(data['date'])

    # Extract the year from the 'date' column
    data['Year'] = data['date'].dt.year

    # Create subplots for each 'Assigned_Issue' with two plots per line
    assigned_issues = data[issue_column].unique()
    num_plots = len(assigned_issues)

    # Define the number of columns per row
    columns_per_row = 2

    # Calculate the number of rows required
    num_rows = (num_plots + columns_per_row - 1) // columns_per_row

    for i, issue in enumerate(assigned_issues):
        # Create a new subplot for every two plots
        if i % columns_per_row == 0:
            plt.figure(figsize=(12, 4))

        # Filter data for the current 'Assigned_Issue'
        issue_data = data[data[issue_column] == issue]

        # Get unique years available in the dataset
        unique_years = issue_data[year_column].unique()

        # Sort unique years to ensure a proper line plot
        unique_years.sort()

        # Create a line plot with markers and a line that passes through each marker
        plt.subplot(1, 2, i % columns_per_row + 1)  # Subplot for two plots per line
        sns.lineplot(data=issue_data, x=year_column, y=rating_column, marker='o', ci=None, sort=False)

        plt.title(f'Distribution of Stars Rating for {issue}')
        plt.xlabel('Year')
        plt.ylabel('Average Stars Rating')

        # Show only x-axis values for years available in the dataset
        plt.xticks(unique_years)

        # Tight layout for better spacing
        plt.tight_layout()

        # If this is the last plot in the row, show the figure
        if (i + 1) % columns_per_row == 0 or i == num_plots - 1:
            plt.show()

generate_star_rating_line_plots(neg_df, 'Assigned_Issue', 'Year', 'stars')
```

### Issues Related to  Customer Service & Support

```python
def generate_summary(reviews):
    # Initializing summarizer --> this will summarize text
    summarizer = pipeline("summarization", model="lidiya/bart-base-samsum")

    # Convert Timestamp objects to strings
    reviews = [str(review) for review in reviews.flatten()]

    # Join the reviews into a single string
    strr = " ".join(reviews)

    # Summarize the concatenated text
    s1 = summarizer(strr[0:3000], min_length=200, max_length=400)
    
    # Print the summary
    print('Summary:')
    print(s1[0]['summary_text'])
```

# Recommendations based on Assigned Issue

In this section, a system is established to analyze airline service reviews. This involves defining issue keywords and related recommendations,  functions to detect issues and generate suggestions from review text to improved customer feedback analysis.

#### Creating new column in df for recommendations

```python
issue_recommendations = {
    'Check-in & Boarding': "To improve the check-in and boarding experience, focus on streamlining the check-in process, providing clear instructions, and ensuring efficient boarding procedures.",
    'Flight Delays & Cancellations': "In the event of flight delays or cancellations, prioritize effective communication with passengers, offer alternative travel options promptly, and provide compensation when appropriate.",
    'Customer Service & Support': "Enhance customer service by training staff to be more empathetic and responsive to passenger concerns, and establish a more efficient and responsive customer support system.",
    'Baggage Issues': "To minimize baggage-related problems, improve baggage handling procedures, invest in better tracking systems, and expedite the process of locating and delivering lost baggage.",
    'Food & Catering': "Focus on improving the quality and variety of in-flight meals, catering to diverse dietary preferences, and ensuring that passengers have a pleasant dining experience.",
    'Seating & Comfort': "Enhance passenger comfort by providing more spacious seating options, offering comfortable amenities, and ensuring that passengers have an enjoyable and relaxing journey.",
    'Entertainment & Amenities': "Enhance the in-flight entertainment experience by expanding entertainment options, providing updated technology, and ensuring that passengers have access to a wide range of amenities.",
    'Cleanliness & Hygiene': "Maintain high standards of cleanliness and hygiene by implementing rigorous cleaning and sanitation procedures on aircraft and within airport facilities.",
    'Security & Safety': "Continue to prioritize security and safety measures to ensure that passengers feel secure throughout their journey and comply with all safety protocols.",
    'Booking & Refund Issues': "Improve the booking and refund process by offering transparent policies, simplifying reservation procedures, and providing efficient solutions for refund requests."
}


def identify_issues_and_generate_suggestions(review_text,issues_keywords,issue_recommendations):
    identified_issues = []
    suggestions = []

    # Check the review text for keywords related to issues
    for issue, keywords in issues_keywords.items():
        for keyword in keywords:
            if keyword in review_text.lower():
                identified_issues.append(issue)
                suggestions.append(issue_recommendations.get(issue, "No specific recommendation available."))

    return identified_issues, suggestions


def generate_recommendations(row):
    identified_issues, suggestions = identify_issues_and_generate_suggestions(
        row['reviews'], issues_keywords, issue_recommendations)
    return ', '.join(suggestions)  # Combine multiple recommendations into one string


neg_df['recommendations'] = neg_df.apply(generate_recommendations, axis=1)
neg_df
```

#### Identifying Issues and Generating Recommendation on new Review

```python
#paste review inside the string review_text
review_text = "✅ Trip Verified | My daughter and I were denied boarding on our business class flight from Madrid to Heathrow in March 25, 2023..."
identified_issues, suggestions = identify_issues_and_generate_suggestions(review_text,issues_keywords,issue_recommendations)

# Display the identified issues and improvement suggestions
if identified_issues:
    print("Identified Issues:", identified_issues)
    print("Improvement Suggestions:")
    for i, suggestion in enumerate(suggestions):
        print(f"{i + 1}. {suggestion}")
else:
    print("No specific issues identified in the feedback.")
```
